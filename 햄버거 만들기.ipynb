```python
from collections import deque

def solution(ingredient):
    ingredient = deque(ingredient)
    burger = []
    answer = 0
    
    while ingredient:
        burger.append(ingredient.popleft())
        if burger[-4:] == [1,2,3,1]:
            burger = burger[:-4]
            answer+=1
            
    return answer
```

```python
from collections import deque

def solution(ingredient):
    ingredient = deque(ingredient)
    burger = []
    answer = 0
    
    while ingredient:
        burger.append(ingredient.popleft())
        if burger[-4:] == [1,2,3,1]:
            for _ in range(4):
                burger.pop()
            answer+=1        
    return answer
```
